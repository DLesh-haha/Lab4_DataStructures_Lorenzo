```python
# EXERCISE 1

# Required Variables
SELF_NUM = "LORENZO"
STUDENT_MAJOR = "BSECE"
FAVORITE_ARTIST = "Laufey"
LOCAL_HAZARD = "Fire"

CONTROL_NUM = max(3, len(SELF_NUM))

from collections import defaultdict, Counter

# Simulated stream
stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]

# Data Expansion
stream.extend([
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", STUDENT_MAJOR),
    ("Visayas", f"BSME_{CONTROL_NUM}"),
    ("TUP Manila", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("Visayas", "BSECE")
])

# Aggregation
campus_programs = defaultdict(list)
for campus, program in stream:
    campus_programs[campus].append(program)

# Frequency Analysis
campus_frequency = {}
for campus in campus_programs:
    campus_frequency[campus] = Counter(campus_programs[campus])

# Most common program
top_program = campus_frequency["TUP Manila"].most_common(1)

# ===== ASSESSMENT DATA =====
total_manila = sum(campus_frequency["TUP Manila"].values())
stream_length = len(stream)

top_program_name = top_program[0][0]
top_program_freq = top_program[0][1]

print("===== ASSESSMENT DATA =====")
print("CONTROL_NUM Used:", CONTROL_NUM)
print("STUDENT_MAJOR Used:", STUDENT_MAJOR)
print("Expanded Stream Length:", stream_length)
print("Total Applications Processed for TUP Manila:", total_manila)
print("Top Requested Program at TUP Manila:", top_program_name)
print("Exact Frequency of Top Program:", top_program_freq)
```

```python
# EXERCISE 2

# Required Variables
SELF_NUM = "LORENZO"
FAVORITE_ARTIST = "Laufey"

CONTROL_NUM = max(3, len(SELF_NUM))

# Festival lineup
festival = {
    "Ben&Ben", "SB19", "Bini", "Eraserheads",
    FAVORITE_ARTIST, "Zild", f"Indie Artist {CONTROL_NUM}"
}

# User Profiles
user_a = {
    "Ben&Ben", "Bini", "Maki", "Dionela", FAVORITE_ARTIST
}

user_b = {
    "SB19", "Eraserheads", "Zild",
    f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"
}

# Data Expansion
festival.update({
    "Arthur Nery",
    "December Avenue",
    "Unique Salonga"
})

# Jaccard Similarity for User A
intersection_a = user_a.intersection(festival)
union_a = user_a.union(festival)
jaccard_a = (len(intersection_a) / len(union_a)) * 100

# Jaccard Similarity for User B
intersection_b = user_b.intersection(festival)
union_b = user_b.union(festival)
jaccard_b = (len(intersection_b) / len(union_b)) * 100

# Dealbreakers for User A
dealbreakers = user_a - festival

# ===== ASSESSMENT DATA =====
print("===== ASSESSMENT DATA =====")
print("Total Artists in Expanded Festival:", len(festival))
print("Length of Intersection (User A & Festival):", len(intersection_a))
print("Length of Union (User A & Festival):", len(union_a))
print("User A Jaccard Similarity (%):", jaccard_a)
print("User B Jaccard Similarity (%):", jaccard_b)
print("User A Dealbreaker Artists:", dealbreakers)
```

```python
# EXERCISE 3

from collections import deque, Counter

# Required variables (edit if needed)
SELF_NUM = "LORENZO"
LOCAL_HAZARD = "FIRE"

CONTROL_NUM = max(3, len(SELF_NUM))

# Initialize deque with max length 5
window = deque(maxlen=5)

# Base burst data
burst = [
    (CONTROL_NUM, "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, LOCAL_HAZARD),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]

# Data Expansion (add CONTROL_NUM + 2 custom alerts)
start_time = CONTROL_NUM + 7

for i in range(CONTROL_NUM + 2):
    burst.append((start_time + i, "WEATHER"))  # pwede mo palitan category

# Process sliding window
accepted = []
weather_plus3_appended = "N"
local_hazard_appended = "N"

for timestamp, category in burst:
    
    # Get current categories in window
    categories = [c for (_, c) in window]
    freq = Counter(categories)
    
    # Spam rule: reject if category already appears 2 or more times
    if freq[category] >= 2:
        continue
    
    # Accept alert
    window.append((timestamp, category))
    accepted.append((timestamp, category))
    
    # Track required answers
    if timestamp == CONTROL_NUM + 3 and category == "WEATHER":
        weather_plus3_appended = "Y"
    
    if category == LOCAL_HAZARD:
        local_hazard_appended = "Y"

# Compute final sum of timestamps
final_sum = sum([t for (t, _) in window])

# OUTPUTS (for assessment table)
print("LOCAL_HAZARD Used:", LOCAL_HAZARD)
print("Total Burst Size:", len(burst))
print("Did WEATHER at CONTROL_NUM + 3 append?:", weather_plus3_appended)
print("Did LOCAL_HAZARD append?:", local_hazard_appended)
print("Final Deque Length:", len(window))
print("Final Sum of Timestamps:", final_sum)
print("Exact Final Deque Output:", list(window))
```
